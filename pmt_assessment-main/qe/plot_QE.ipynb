```python
#import plotly.express as px 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv_ref= pd.read_csv('./Sensitivity_H11901-04.csv')
wavelength_ref = csv_ref['x'].values #[nm]
sensitivity_ref = csv_ref['y'].values #[mA/W]
qe_ref = (sensitivity_ref*1e-3 * 1240)/wavelength_ref * 100 #%
```

```python
csv_ham= pd.read_csv('./Sensitivity_R14374-31.csv')
wavelength_ham = csv_ham['x'].values #[nm]
sensitivity_ham = csv_ham['y'].values #[mA/W]
qe_ham = (sensitivity_ham*1e-3 * 1240)/wavelength_ham * 100 #%
```

```python
csv_ham2= pd.read_csv('./QE_R14374-31.csv')
wavelength_ham2 = csv_ham2['x'].values #[nm]
qe_ham2 = csv_ham2['y'].values #%
```

```python
plt.plot(wavelength_ref, qe_ref, label='H11901P-04')
plt.plot(wavelength_ham, qe_ham, label='R14374-31')
plt.plot(wavelength_ham2, qe_ham2, label='R14374-31 (org)')
plt.xlabel('Wavelength [nm]')
plt.ylabel('Quantum Efficiency %')
plt.yscale('log')
plt.ylim([1e-1,1e2])
plt.xlim([300,800])
plt.grid()
plt.legend()
plt.show()
```
